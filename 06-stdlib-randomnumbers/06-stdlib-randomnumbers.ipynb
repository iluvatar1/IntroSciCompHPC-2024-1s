## Special functions
https://en.cppreference.com/w/cpp/numeric/special_functions

### Gamma function
See <http://en.cppreference.com/w/cpp/numeric/math/tgamma>

``` cpp
#include <cstdio>
#include <cmath>

int main(void)
{
  const double XMIN = 0.0;
  const double XMAX = 11.0;
  const double DX = 0.1;
  const int NSTEPS = int((XMAX-XMIN)/DX);

  for(int ii = 0; ii < NSTEPS; ++ii) {
    double x = XMIN + ii*DX;
    printf("%25.16e%25.16e\n", x, std::tgamma(x));
  }

  return 0;
}
```

### Beta function

``` cpp
// requires c++17
#include <cstdio>
#include <cmath>

int main(void)
{
  const double XMIN = -3.0;
  const double XMAX = 3.0;
  const double YMIN = -3.0;
  const double YMAX = 3.0;
  const double DELTA = 0.01;
  const int NSTEPS = int((XMAX-XMIN)/DELTA);

  for(int ii = 0; ii < NSTEPS; ++ii) {
    double x = XMIN + ii*DELTA;
    for(int jj = 0; jj < NSTEPS; ++jj) {
      double y = YMIN + jj*DELTA;
      printf("%25.16e%25.16e%25.16e\n", x, y, std::beta(x, y));
    }
    printf("\n");
  }

  return 0;
}
```

### Uniform distribution with random seed
<div style="text-align: center;">
    <img src="https://hackingcpp.com/cpp/std/random/uniform_real_distribution_plot.svg" alt="Image Description" width="600">
    <figcaption>From: "https://hackingcpp.com/cpp/std/random/uniform_real_distribution_plot.svg"</figcaption>

``` cpp
#include <random>
#include <iostream>

int main(void)
{
  //std::random_device rd; // inicializacion con semilla aleatoria
  //std::mt19937 gen(rd()); // genera bits aleatorios
  std::mt19937 gen(10);
  std::uniform_real_distribution<> dis(1, 2); // distribucion
  for(int n = 0; n < 10; ++n) {
    std::cout << dis(gen) << std::endl;
  }
}
```

### Uniform distribution controlled seed

``` cpp
#include <random>
#include <iostream>
#include <cstdlib>

int main(int argc, char **argv)
{
  int seed = std::atoi(argv[1]);
  std::mt19937 gen(seed);
  std::uniform_real_distribution<double> dis(1, 2);
  for(int n = 0; n < 100000; ++n) {
    std::cout << dis(gen) << std::endl;
  }
  return 0;
}
```

### Normal distribution controlled seed
